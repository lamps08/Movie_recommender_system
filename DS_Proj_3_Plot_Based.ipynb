```python
import pandas as pd
```

```python
inp_data = pd.read_csv('movie_metadata.csv')
print inp_data.columns
inp_data.head()
```

```python
data = inp_data[[u'movie_title', u'genres', u'plot_keywords']].dropna()
```

```python
data.head()
```

```python
names = data.movie_title.values
print names[:2]
genres = data.genres.map(lambda x: x.lower().replace(" ", "").replace("|"," ")).values
print genres[:2]
plot = data.plot_keywords.map(lambda x: x.lower().replace("|"," ")).values
print plot[:2]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(stop_words='english')
genre_matrix = vectorizer.fit_transform(genres)
print genre_matrix.shape
```

```python
vectorizer = CountVectorizer(stop_words='english')
plot_matrix = vectorizer.fit_transform(plot)
print plot_matrix.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.metrics.pairwise import manhattan_distances
from sklearn.metrics import jaccard_similarity_score
```

## For Genres Similarity

```python
inp= genre_matrix[3]
print names[3], genres[3]
```

```python
data_genre_sim = []
for idx in range(genre_matrix.shape[0]):
    data_genre_sim.append((names[idx], genres[idx], cosine_similarity(inp, genre_matrix[idx])[0][0], \
                           jaccard_similarity_score(inp, genre_matrix[idx]),\
                           euclidean_distances(inp, genre_matrix[idx])[0][0],\
                          manhattan_distances(inp, genre_matrix[idx])[0][0] ))
```

```python
pd.DataFrame(data_genre_sim, columns=['name', 'genres', 'cos', 'jac', 'euc', 'man']).sort_values(by='man').head(50)
```

## For Plot Similarity

```python
inp= plot_matrix[3]
print names[3], plot[3]
```

```python
data_plot_sim = []
for idx in range(plot_matrix.shape[0]):
    data_plot_sim.append((names[idx], plot[idx],\
                          cosine_similarity(inp, plot_matrix[idx])[0][0],\
                          #jaccard_similarity_score(inp, plot_matrix[idx]),\
                          euclidean_distances(inp, plot_matrix[idx])[0][0]))
```
